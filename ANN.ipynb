```python
X=df.drop(['Exited'],axis=1)
X=pd.get_dummies(X,drop_first=True,dtype='int')
y=df['Exited']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
import warnings
warnings.filterwarnings('ignore')

from keras.layers import Dense
from keras.models import Sequential
seq=Sequential()
```

```python
seq.add(Dense(input_dim =11, units=12, kernel_initializer='uniform',activation = 'relu'))
seq.add(Dense(units=11,kernel_initializer='uniform',activation ='relu'))
seq.add(Dense(units=10,kernel_initializer='uniform',activation ='relu'))

seq.add(Dense(units=1,kernel_initializer='uniform',activation ='sigmoid'))
          
```

```python
import warnings 
warnings.filterwarnings('ignore')

seq.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# seq.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuray'])
```

```python
X_train
```

```python
y_train
```

```python
seq.fit(X_train,y_train,epochs=100,batch_size=32,verbose=1)
```

```python
y_pred=seq.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print('test Accuracy:',accuracy_score(y_test,y_pred))
confusion_matrix(y_test,y_pred)
```
